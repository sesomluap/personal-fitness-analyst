```python

# https://api.census.gov/data/2020/acs/acs5?get=NAME,B17001_002E,B17001_003E,B17001_017E&for=county:053&in=state:27

# Initialize an empty DataFrame
df = pd.DataFrame()

# Loop over years
for year in range(2005, 2023):
    # Specify the county FIPS code
    hennepin_county_code = '053'
    
    # Construct the API URL with the API key
    url = f'https://api.census.gov/data/{year}/acs/acs5?get=NAME,B17001_002E,B17001_003E,B17001_017E&for=county:{hennepin_county_code}&in=state:27&key={census_key}'
    print(f'URL: {url}')
    # Make the API request
    response = requests.get(url)
    
    if response.status_code == 200:
        print(f"Year: {year}, County FIPS: {hennepin_county_code}, Response Content: {response.text}")
        
        # Convert the response to a DataFrame
        county_data = pd.DataFrame(response.json()[1:], columns=response.json()[0])
        
        # Append the year and county columns to the DataFrame
        county_data['Year'] = year
        county_data['County_FIPS'] = hennepin_county_code
        
        # Append the county data to the main DataFrame
        df = df.append(county_data, ignore_index=True)
    else:
        print(f"Failed to fetch data for {year} - County FIPS {hennepin_county_code}. Status Code: {response.status_code}")

# Display the head of the resulting DataFrame
print(df.head())
```
